1. Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.


```python
numbers = []
for i in range(2000,3201):
  if i % 7 == 0 and i % 5 != 0:
    numbers.append(i)
print(numbers)
```

2. Write a function that get a matrix as input and returns normalized matrix: normalization formula: (x-mean/std)

```python
import numpy as np

def normalization(matrix):
    mean = np.mean(matrix)
    std = np.std(matrix)
    return (matrix - mean) / std

matrix = np.array([[1, 2], [3, 4]])
normalized_matrix = normalization(matrix)
print(normalized_matrix)
```

3. Write a function that get a float number and returns integer number and decimal.


```python
def integer_decimal(x):
  integer = int(x)
  decimal = x - integer
  return integer, decimal
integer_decimal(3.4)
```

4. Write a program that accepts a sentence and calculate the number of upper case letters and lower case letters.
Suppose the following input is supplied to the program:
Hello world!
Then, the output should be:
UPPER CASE 1
LOWER CASE 9

```python
x = "Hello world!"
upper_case = 0
lower_case = 0
for i in x:
  if i.isupper():
    upper_case += 1
  if i.islower():
    lower_case += 1
print("UPPER CASE",upper_case)
print("LOWER CASE",lower_case)
```

5. A website requires the users to input username and password to register. Write a program to check
the validity of password input by users.

Following are the criteria for checking the password:

At least 1 letter between [a-z]

At least 1 number between [0-9]

At least 1 letter between [A-Z]

At least 1 character from [$#@]

Minimum length of transaction password: 6

Maximum length of transaction password: 12

Your program should accept a sequence of comma separated passwords and will check them
according to the above criteria. Passwords that match the criteria are to be printed, each separated
by a comma.

Example

If the following passwords are given as input to the program:
ABd1234@1,a F1#,2w3E*,2We3345

Then, the output of the program should be:
ABd1234@1

Hints:
In case of input data being supplied to the question, it should be assumed to be a console input.

```python
import time
approved = []
def check_password(password):
    if len(password) < 6 or len(password) > 12:
        return False
    has_upper_letter = False
    has_lower_letter= False
    has_digit = False
    has_special_character = False
    for i in password:
        if i.islower():
            has_lower_letter = True
        elif i.isdigit():
            has_digit = True
        elif i.isupper():
            has_upper_letter = True
        elif i in "$#@":
            has_special_character = True
    if has_upper_letter and has_lower_letter and has_digit and has_special_character:
        return True
   
passwords = "ABd1234@1,a F1#,2w3E*,2We3345".split(",")
print("We are now checking your passwords..")
time.sleep(3)

for j in passwords:
    if check_password(j):
        approved.append(j)
print(approved)
```
